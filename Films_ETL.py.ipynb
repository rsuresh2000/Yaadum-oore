```python
import requests, bs4
import pandas as pd

filmFile = open('allfilms_2010s.txt','w')
castFile = open('allstars_2010s.txt','w')

filmList = []
castList = []

urlpath = 'https://en.wikipedia.org/wiki/List_of_Tamil_films_of_'
yr=0

for yr in range(2010,2019,1):
    url=urlpath+str(yr)  
      
    try:
        resp = requests.get(url).content
        bsresp = bs4.BeautifulSoup(resp)
    except:
        continue
    else:
        cast = bsresp.select('tr > td > a:nth-of-type(2)')
        cols =[x.get_text(strip=True) for x in cast]
        cols=[x.encode('utf-8') for x in cols]
        colset=set(cols)
        cols=sorted(colset)
        castList.extend(cols)
        
        film = bsresp.select('tr > td > i:nth-of-type(1)')
        cols =[x.get_text(strip=True) for x in film]
        cols=[x.encode('utf-8') for x in cols]
        colset=set(cols)
        cols=sorted(colset)
        filmList.extend(cols)

allCast=sorted(set(castList))
allFilm=sorted(set(filmList))

for item in allFilm:
    filmFile.write("%s\n" % item)
    
for item in allCast:
    castFile.write("%s\n" % item)  
```

```python
import sqlite3
import os

def create_tbl(conn):
    sqlfilms = '''create table if not exists Films(
            filmid INTEGER PRIMARY KEY,
            title TEXT,
            released INTEGER,
            length TEXT,
            genre TEXT,
            reception TEXT,
            myrating REAL,
            plot TEXT,
            filmid_uniq INTEGER);'''
    
    cur.execute(sqlfilms)
    
    sqlartists = '''create table if not exists Artists(
            artistid INTEGER PRIMARY KEY,
            artist TEXT,
            gender TEXT,
            age INTEGER,
            info TEXT); '''
    
    cur.execute(sqlartists)
    
    sqlfilmartist = '''create table if not exists Film_Artist(
            filmid INTEGER,
            artistid INTEGER,
            roleid INTEGER,
            FOREIGN KEY(filmid) REFERENCES Films(filmid),
            FOREIGN KEY(roleid) REFERENCES Roles(roleid),
            FOREIGN KEY(artistid) REFERENCES Artists(artistid)); '''

    cur.execute(sqlfilmartist)
    
    sqlroles = '''create table if not exists Roles(
             roleid INTEGER PRIMARY KEY,
             role TEXT,
             role_name TEXT); '''

    cur.execute(sqlroles)
        
    
#     sqlsongs = '''create table if not exists Songs(
#             songid INTEGER PRIMARY KEY,
#             filmid INTEGER,
#             songlist TEXT
#             FOREIGN KEY (filmid) REFERENCES Films(filmid));'''


conn = sqlite3.connect('films.db')
cur = conn.cursor()
create_tbl(conn)

print ("Created Films table successfully ")


# sql = '''SELECT * FROM sqlite_master WHERE type='table' ORDER BY name;'''
# tbls=cur.execute(sql)
# print(tbls)
```
